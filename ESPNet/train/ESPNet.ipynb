```
import os
import glob


def main():
    os.chdir('./data_road/Road_data/')
    Create_txt(train_txt,0)
    Create_txt(val_txt,1)
    os.chdir('/content/drive/My Drive/Colab Notebooks/ESPNet/train')
    

def Create_txt(txt_path,j):
    filelist_rgb = glob.glob(os.path.join('*.jpg'))
    txt = os.path.basename(txt_path)
    i=0
    try:
        with open(txt_path, mode='x') as f:
            for file_path_rgb in filelist_rgb:
              if j == 0:
                if i%10 != 0:
                  s = 'Road_data/' + file_path_rgb + ',Road_label/' + file_path_rgb.strip('.jpg') + '.png' + '\n'
                  f.write(s)
                i+=1
              else: 
                if i%10 ==0:
                  s = 'Road_data/' + file_path_rgb + ',Road_label/' + file_path_rgb.strip('.jpg') + '.png' + '\n'
                  f.write(s)
                i+=1


        print ('create :', txt)

    except FileExistsError:
        print(txt, 'already exists')

if __name__ == '__main__':
    #書き換えが必要です

    #jisakuの部分は好きな名前で構いません
    #※プログラムを実行する前にディレクトリを作成しておいてください
    train_txt = '/content/drive/My Drive/Colab Notebooks/ESPNet/train/data_road/train.txt'
    val_txt = '/content/drive/My Drive/Colab Notebooks/ESPNet/train/data_road/val.txt'

    main()
```
